# Demonstration 1.3.1 Normal and binomial distribution

Follow the demonstration to learn about normal and binomial distributions. In this demonstration, you will learn:
- What normal and binomial distributions are.
- How to create a normal and binomial distribution.
- How to interpret a normal and binomial distribution.

## a. Normal distribution: Overview
The normal distribution, also known as the Gaussian distribution, is a common continuous probability distribution. It is often used to represent real-valued random variables whose distributions are not known. A classic example is the height of adults in a population, which usually forms a bell curve around the average height.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Normal distribution simulation:
# Mean and standard deviation
mean, std_dev = 0, 1
normal_sample = np.random.normal(mean, std_dev, 1000)
```

```python
# Plot the normal distribution.
plt.figure(figsize=(8, 6))
sns.histplot(normal_sample, kde=True, color='blue')
plt.title('Normal Distribution (Mean=0, Std Dev=1)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## b. Exploring the normal distribution
By changing the mean and standard deviation, we can see how the shape of the normal distribution changes. A larger standard deviation stretches the curve, while a smaller one compresses it. The mean shifts the center of the curve left or right.


```python
# Import interact.
from ipywidgets import interact

# Function to update the plot.
def update_plot(mean, std_dev):
    plt.figure(figsize=(8, 6))
    normal_sample = np.random.normal(mean, std_dev, 1000)
    sns.histplot(normal_sample, kde=True, color='blue')
    plt.title(f'Normal Distribution (Mean={mean}, Std Dev={std_dev})')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()

# Interactive slider.
interact(update_plot, mean=(-5, 5, 0.5), std_dev=(0.1, 5, 0.1))
```

## c. Binomial distribution: Overview
The binomial distribution is a discrete probability distribution. It describes the outcome of binary scenarios, e.g. flipping a coin. It is characterised by two parameters: the number of trials (n) and the probability of success (p) in each trial.


```python
# Binomial distribution simulation:
# Number of trials and probability of success.
n, p = 10, 0.5
binomial_sample = np.random.binomial(n, p, 1000)
```

```python
# Plot the binomial distribution.
plt.figure(figsize=(8, 6))
sns.histplot(binomial_sample,
             bins=10, kde=False,
             color='green')
plt.title('Binomial distribution (n=1000, p=0.5)')
plt.xlabel('Number of successes')
plt.ylabel('Frequency')
plt.show()
```

## d. Understanding binomial distribution
In this interactive section, adjust the number of trials and the probability of success. Observe how the distribution changes. This is particularly useful in understanding phenomena with binary outcomes, like success/failure, yes/no, etc.


```python
# Function to update the binomial distribution plot.
def update_binomial_plot(n, p):
    plt.figure(figsize=(8, 6))
    binomial_sample = np.random.binomial(n, p, 1000)
    sns.histplot(binomial_sample, bins=n, kde=False, color='green')
    plt.title(f'Binomial Distribution (n={n}, p={p})')
    plt.xlabel('Number of Successes')
    plt.ylabel('Frequency')
    plt.show()

# Interactive sliders for number of trials and probability of success.
interact(update_binomial_plot, n=(1, 100, 1), p=(0, 1, 0.05))
```
